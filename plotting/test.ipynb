```python
import numpy as np
import matplotlib.pyplot as plt

# Load data (skip the header row)
data = np.loadtxt("./trajectory.csv", delimiter=",", skiprows=1)

# Split into columns
t, x, y, z = data[:, 0], data[:, 1], data[:, 2], data[:, 3]

# Plot in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot(x, y, z, lw=0.8)

ax.set_title("Lorenz Attractor")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.tight_layout()
plt.show()
```

```python
# Define the Lorenz system
def lorenz(t, y, sigma=10.0, beta=8.0/3.0, rho=28.0):
    x, y_, z = y
    dx = sigma * (y_ - x)
    dy = x * (rho - z) - y_
    dz = x * y_ - beta * z
    return np.array([dx, dy, dz])

# Runge-Kutta 4th order integrator
def rk4_step(f, t, y, dt):
    k1 = f(t, y)
    print(k1)
    k2 = f(t + dt / 2, y + dt * k1 / 2)
    print(k2)
    k3 = f(t + dt / 2, y + dt * k2 / 2)
    print(k3)
    k4 = f(t + dt, y + dt * k3)
    print(k4)
    return y + dt * (k1 + 2*k2 + 2*k3 + k4) / 6

y0 = np.array([10.0, 10.0, 10.0])
t0 = 0.0
dt = 0.01

print(rk4_step(lorenz, t0, y0, dt))
```
